```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = 'Resources/purchase_data.csv'
heroes_df = pd.read_csv(csv_path)

heroes_df.head()
```

```python
heroes_df.describe()
```

```python
Total_Revenue = heroes_df["Price"].sum()
Total_Revenue
```

```python
Average_Purchase_Price = heroes_df["Price"].mean()
Average_Purchase_Price
```

```python
Total_Number_Purchases = heroes_df["Purchase ID"].count()
Total_Number_Purchases
```

```python
Total_Number_Players = heroes_df["SN"].value_counts()
Total_Number_Players
```

```python
Number_of_Unique_items = heroes_df["Item Name"].value_counts()
Number_of_Unique_items
```

```python
count_gender = heroes_df["Gender"].value_counts()
count_gender
```

```python
grouped_gender_df = heroes_df.groupby(["Gender"])
print(grouped_gender_df)
grouped_gender_df.count().head(10)
```

```python
Total_Purchase_Count_Gender = grouped_gender_df["Purchase ID"].count()
Total_Purchase_Count_Gender.head()
```

```python
Total_Average_PurchasePrice_Gender = grouped_gender_df["Price"].mean()
Total_Average_PurchasePrice_Gender.head()
```

```python
Total_Purchase_Value_Gender = grouped_gender_df["Price"].sum()
Total_Purchase_Value_Gender.head()
```

```python
summary_gender_df = pd.DataFrame({"Purchase Count": Total_Purchase_Count_Gender,
                                  "Average Purchase Price": Total_Average_PurchasePrice_Gender,
                                  "Total Purchase Value": Total_Purchase_Value_Gender})
summary_gender_df.head()
```

```python
grouped_gender_person_df = heroes_df.groupby(["Gender","SN"])
print(grouped_gender_person_df)
```

```python
Average_Purchase_Person_Gender_df = pd.DataFrame(grouped_gender_person_df["Price"].mean())
Average_Purchase_Person_Gender_df.head(10)
```

```python
bins = [0,9,14,19,24,29,34,40]
group_ages = ["<10","10-14","15-19","20-24","25-29","30-34","35-40"]
```

```python
pd.cut(heroes_df["Age"], bins, labels=group_ages).head()
```

```python
heroes_df["Age Group"] = pd.cut(heroes_df["Age"], bins, labels=group_ages)
heroes_df.head()
```

```python
age_group = heroes_df.groupby("Age Group")
print(age_group["SN"].count())
```

```python
age_group[["Item ID"]].count()
```

```python
age_group[["Price"]].mean()
```

```python
age_group[["Price"]].sum()
```

```python
spenders_df = heroes_df[["SN","Price"]]
spenders_df.head(10)
```

```python
spenders_grouped = spenders_df.groupby(["SN"])
print(spenders_grouped)
spenders_comparison_df = [[spenders_grouped.mean()],[spenders_grouped.sum],[spenders_grouped.count()]]
spenders_comparison_df
```

```python
grouped_SN_df = heroes_df.groupby(["SN"])
print(grouped_SN_df)
grouped_SN_df.head(10)
```

```python
grouped_bySN_df = pd.DataFrame(grouped_SN_df["Price"].sum())
grouped_bySN_df.head(5)
```

```python
top_spenders_df = grouped_bySN_df.sort_values("Price",ascending=False)
top_spenders_df.head(5)
```

```python
grouped_item_df = heroes_df.groupby(["Item Name"])
print(grouped_item_df)
grouped_item_df.head(20)
```

```python
item_purchases_df = pd.DataFrame(grouped_item_df["Item ID"].count())
item_purchases_df.head(10)
```

```python
top_popular_df = item_purchases_df.sort_values("Item ID",ascending=False)
top_popular_df.head(5)
```

```python
item_profit_df = pd.DataFrame(grouped_item_df["Price"].sum())
item_profit_df.head(10)
```

```python
top_profit_df = item_profit_df.sort_values("Price", ascending=False)
top_profit_df.head(5)
```
